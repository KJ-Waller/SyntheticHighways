```python

import pyproj
geodesic = pyproj.Geod(ellps='WGS84')

class OUActionNoise(object):
    def __init__(self, mu, sigma=3.0, theta=1.0, dt=.0001, x0=None):
        self.theta = theta
        self.mu = mu
        self.sigma = sigma
        self.dt = dt
        self.x0 = x0
        self.reset()

    def __call__(self):
        x = self.x_prev + self.theta * (self.mu - self.x_prev) * self.dt + \
            self.sigma * np.sqrt(self.dt) * np.random.normal(size=self.mu.shape)

        self.x_prev = x
        return x

    def reset(self):
        self.x_prev = self.x0 if self.x0 is not None else np.zeros_like(self.mu)
        
def init_noisemodel(meters=5.0):
    deg = m_to_deg(meters)
    
    mu = np.array([deg,deg])
    x0 = np.array([deg,deg])
#     noise_sigma, noise_theta = 1.25e-05, 1.25e-05
#     noise_sigma, noise_theta = deg*1e-5, deg*1e-5
    noise_sigma, noise_theta = 1e-5, 1e-5
    noise_dt = .1
    
    noise_function = OUActionNoise(mu, noise_sigma, noise_theta, noise_dt)
    
    return noise_function


def m_to_deg(meters, x0=[0,0], x1=[1.30485874e+02, 1.41421356e-03]):
    
    deg = np.interp(meters, x0, x1)
    return deg

def add_noise_t(t, meters=10):
    t = t.copy()
    
    noise_function = init_noisemodel(meters=meters)
    
    
    ou_noise = np.array([noise_function() for i in range(len(t))])
    old_coords = np.array([(p['lat'], p['lon']) for p in t])
    new_coords = old_coords + ou_noise
    
    plt.plot(np.arange(0, ou_noise.shape[0]), ou_noise)

    final_noise = new_coords

    t['lat'], t['lon'] = final_noise[:,0], final_noise[:,1]

    noise_function.reset()

    return t

def measure_noise(t, t_):
    t = np.stack([t['lat'].view('f8'), t['lon'].view('f8')], axis=1)
    t_ = np.stack([t_['lat'].view('f8'), t_['lon'].view('f8')], axis=1)
    _, _, dists = geodesic.inv(t[:,1], t[:,0], t_[:,1], t_[:,0])
    t_diff = t - t_
    t_diff = np.linalg.norm(t_diff, axis=1)
    return dists, t_diff

def plot_noise_t(G1, t):
    plot_graph(snapshot_to_nxgraph(G1, [t]), zoom_on_traj=True, show_nodes=True, T_node_size=5.0, figsize=(7,7))
    
    t_ = add_noise_t(t,meters=100)
    plot_graph(snapshot_to_nxgraph(G1, [t_]), zoom_on_traj=True, show_nodes=True, T_node_size=5.0, figsize=(7,7))
    
    dists, t_diff = measure_noise(t,t_)
    plt.plot(np.arange(0,dists.shape[0]), dists)
    plt.xlabel('timed points')
    plt.ylabel('Distance/noise (m)')
    plt.title(f'Average dist: {np.mean(dists)}')
    
    
plot_noise_t(G1, T2['T'][0])
```

```python
def measure_noise_t(t, t_):
    t = np.stack([t['lat'].view('f8'), t['lon'].view('f8')], axis=1)
    t_ = np.stack([t_['lat'].view('f8'), t_['lon'].view('f8')], axis=1)
    _, _, dists = geodesic.inv(t[:,1], t[:,0], t_[:,1], t_[:,0])
    t_diff = t - t_
    t_diff = np.linalg.norm(t_diff, axis=1)
    return dists, t_diff

def measure_noise():
    dataset_nonoise = SHDataset(noise=False)
    G1,T1,G2,T2 = dataset_nonoise.read_snapshots(0, bbox=(52.34,52.36, 4.90, 4.93))
    
    results = []
    
    for i in range(4):
        dataset = SHDataset(noise=True, noise_config=i)
        G1_,T1_,G2_,T2_ = dataset_nonoise.read_snapshots(0, bbox=(52.34,52.36, 4.90, 4.93))
        
        noise = []
        
        for j in len(T2['T']):
            t = T2['T'][j]
            t_ = T2_['T'][j]
            dists, _ = measure_noise_t(t, t_)
            noise.append(np.mean(dists))
        
        results.append({
            'config': i,
            'noise': np.mean(noise)
        })
        print(f'Dataset noise config {i} has noise {np.mean(noise)}')
```
